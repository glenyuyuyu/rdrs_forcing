```python
from pathlib import Path
import pandas as pd
import xarray as xr
import os
import numpy as np

project_root = Path('/home/glen/works/forcing_generate/')
```

```python
list_paths = []
nc_folder = Path(project_root / 'rdrsv2.1')

for file in sorted(os.listdir(nc_folder)):
    file_path = Path(nc_folder / file)
    list_paths.append(file_path)

# Merge all dates to one dataset
ds = xr.open_mfdataset(list_paths, combine = 'nested', concat_dim="time", parallel=True)

# Save the netcdf file
ds.to_netcdf(project_root / 'RDRSv2.1_UTC0.nc', 'w')

rdrs = xr.open_dataset(project_root / 'RDRSv2.1_UTC0.nc')
rdrs
```

```python
ds = xr.open_dataset(project_root / 'rdrsv2.1' / '1980010112.nc')
ds
```

```python
# time shift
UTC_tz = 5      # UTC-5
old_start_datetime = ds['time'][0].to_numpy()[()]
new_start_datetime = old_start_datetime - np.timedelta64(UTC_tz, 'h')
utc5time = pd.date_range(new_start_datetime, freq='1H', periods=len(ds['time']))
ds['time'] = utc5time
```

```python
# Save the netcdf file
ds.to_netcdf(project_root / 'RDRSv2.1_precip_UTC5_1h.nc', 'w')

rdrs = xr.open_dataset(project_root / 'RDRSv2.1_precip_UTC5_1h.nc')
rdrs
```
